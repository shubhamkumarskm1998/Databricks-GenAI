## Model Context Protocol (MCP)

### Limitations
Generative AI development is complex due to different tools and models.

- Integration issues (MxN problem)
- No common standard → custom work required
- Security challenges in data sharing

### MCP
An open-source standard (by Anthropic) that simplifies AI integration.

- Universal connector (like USB-C for AI tools)
- Easier scaling and faster development
- Secure, controlled access to data and resources

## MCP - Core Components

### MCP Host
The environment where the LLM runs (IDE or app).
Acts as the control center connecting users to tools and data via MCP.

### MCP Client
The bridge between host and servers.
Handles communication and converts requests/responses into MCP format.

### MCP Server
The tool or data source exposed via MCP.
Provides functions or data to the host in a standardized way.

## MCP - Flow

1. **MCP Host** (Claude / IDE / AI App) initiates the request.
2. **MCP Client** acts as a bridge and formats the request.
3. Request is sent via **MCP protocol** to the appropriate **MCP Server**.
4. **MCP Server** interacts with external resources:
   - Local Filesystem
   - Databases
   - Web APIs
5. Server processes the request and returns the response.
6. **MCP Client** converts it back and sends it to the **Host**.
