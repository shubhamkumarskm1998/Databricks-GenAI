### Section 6. Evaluation and Monitoring

### RAG Evaluation

### Evaluation Loop
- Build Evaluation Dataset → Create benchmark Q&A or ground truth data
- Improve RAG Pipeline → Tune retrieval, prompts, embeddings
- Evaluate Pipeline → Measure performance using metrics
- Identify Improvements → Iterate and optimize system

### Key Insight
RAG Evaluation = Continuous feedback loop

---

### What Are We Evaluating?

A RAG system has two critical components:

#### 1. Retrieval
- Are we fetching the right documents?
- Focus → Relevance of retrieved chunks

#### 2. Generation
- Is the LLM producing accurate, complete, and helpful answers?
- Focus → Quality of final response

---

### Key Insight
RAG = Retrieval Quality + Generation Quality

---

### One-line Summary
Evaluate both retrieval and generation in an iterative loop

### Core Evaluation Metrics

### A. Retrieval Evaluation

- Recall@K → % of relevant (gold) documents present in top-K results  
- Precision@K → % of retrieved top-K documents that are actually relevant  
- Hit@K → Did at least one relevant document appear in top-K (Yes/No)  
- Average Similarity → Cosine similarity between query and retrieved vectors  

### Key Insight
Recall = Coverage  
Precision = Accuracy of retrieval  

---

### B. Generation Evaluation

- Factual Consistency → Are answers verifiable from retrieved sources?  
- Completeness → Does the answer fully address the query?  
- Relevance → Is the response aligned with the question?  
- Grounding → Are answers supported by retrieved context (citations)?  
- Toxicity/Bias → Is the output safe and unbiased?  

### Key Insight
Good RAG = Correct + Complete + Grounded + Safe  

---

### One-line Summary
Retrieval metrics evaluate documents, Generation metrics evaluate answers

### Quantitative Evaluation Metrics

### A. Using Synthetic Q&A Pairs
- Create dataset with:
  - Questions
  - Expected answers
  - Relevant document chunks

- Evaluate:
  - Retrieval Accuracy → Was correct document retrieved?
  - Answer Correctness → Does output match expected answer?

### Key Insight
Synthetic data enables automated evaluation without human labeling

---

### B. Embedding Similarity Check
- Use cosine similarity between:
  - Query vector and top-K retrieved vectors
  - Generated response and retrieved chunks

### Key Insight
Higher similarity = Better semantic alignment

---

### C. Automated LLM Grading (LLM-as-a-Judge)
- Use an LLM (e.g., ChatDatabricks) to evaluate responses
- Score based on:
  - Factuality
  - Completeness
  - Relevance

### Key Insight
LLMs can automate large-scale evaluation

---

### D. Human-in-the-Loop Evaluation
- Required for:
  - New domains
  - Regulated use cases
  - Edge cases

### Key Insight
Human evaluation = Highest quality but slow and costly

---

### One-line Summary
Combine synthetic, similarity, LLM, and human methods for full evaluation

### Building an Evaluation Pipeline on Databricks

### Step-by-Step

- Log Responses
  - Store each RAG response in a Delta table
  - Include → query, llm_response, retrieved_chunks, timestamp, prompt_version

- Add Ground Truth
  - Create synthetic dataset
  - Store → expected_answer, expected_chunks

- Compute Metrics (PySpark)
  - Retrieval → Recall@K, Precision@K
  - Similarity → Cosine similarity
  - Generation → BLEU, ROUGE (optional)

- Visualize Metrics
  - Use Databricks SQL / dashboards
  - Track → failures, trends, drift over time

- Manual Review
  - Flag low-confidence responses
  - Send to human review system

---

### Key Insight
Evaluation Pipeline = Logging + Metrics + Monitoring + Human Feedback

---

### One-line Summary
Track, measure, visualize, and review RAG performance continuously

### Perplexity (Foundation Model Metric)

### Definition
Perplexity measures how confident a language model is in predicting the next token.

### Interpretation
- Low Perplexity → High confidence → Better model
- High Perplexity → Low confidence → Poor model

### Intuition
- If the model is sure about the next word → low perplexity
- If the model is confused → high perplexity

### Example
Sentence: "I drink coffee every ___"
- Predicts "morning" with high probability → Low perplexity
- Predicts random words → High perplexity

### Use Cases
- Evaluate language model performance
- Compare different models
- Measure prediction uncertainty

### Limitations
- Focuses on next-token prediction only
- Does not measure answer quality in RAG
- Cannot evaluate factual correctness

### Key Insight
Perplexity = Measure of uncertainty in predictions

### One-line Summary
Lower perplexity means the model is more confident and performs better

### Toxicity (Foundation Model Metric)

### Definition
Toxicity measures how harmful, offensive, or inappropriate a model’s output is.

### Purpose
- Identify and flag harmful language
- Detect offensive or unsafe content
- Ensure safe and responsible AI outputs

### Interpretation
- Low Toxicity → Low harm → Safe output
- High Toxicity → High harm → Unsafe output

### How It Works
- Uses pre-trained classifiers (e.g., hate speech detection models)
- Scores text based on level of harmful content

### Use Cases
- Content moderation
- Safety evaluation of LLM outputs
- Guardrails in GenAI applications

### Key Insight
Toxicity = Measure of safety in generated text

### One-line Summary
Lower toxicity means safer and more acceptable model output

### Inference Table (Databricks)

### What is an Inference Table?
A table that stores every model request and response for monitoring, debugging, and evaluation.

---

### What is Stored?

- Request → User query / input
- Response → Model output
- Retrieved Context → Documents used (RAG)
- Metadata → Timestamp, model version, prompt version

---

### Why It is Important

- Monitoring → Track model performance in production
- Debugging → Analyze failures and incorrect responses
- Evaluation → Compute metrics (accuracy, relevance, latency)
- Auditing → Maintain logs for governance and compliance

---

### Advanced Usage

- Fine-tuning → Use logs to improve model
- Retraining → Use real-world data for better performance
- Drift Detection → Identify changes in data or model behavior

---

### Key Insight
Inference Table = Source of truth for real-world model behavior

---
