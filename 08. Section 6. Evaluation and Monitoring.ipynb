### Section 6. Evaluation and Monitoring

### RAG Evaluation

### Evaluation Loop
- Build Evaluation Dataset → Create benchmark Q&A or ground truth data
- Improve RAG Pipeline → Tune retrieval, prompts, embeddings
- Evaluate Pipeline → Measure performance using metrics
- Identify Improvements → Iterate and optimize system

### Key Insight
RAG Evaluation = Continuous feedback loop

---

### What Are We Evaluating?

A RAG system has two critical components:

#### 1. Retrieval
- Are we fetching the right documents?
- Focus → Relevance of retrieved chunks

#### 2. Generation
- Is the LLM producing accurate, complete, and helpful answers?
- Focus → Quality of final response

---

### Key Insight
RAG = Retrieval Quality + Generation Quality

---

### One-line Summary
Evaluate both retrieval and generation in an iterative loop

### Core Evaluation Metrics

### A. Retrieval Evaluation

- Recall@K → % of relevant (gold) documents present in top-K results  
- Precision@K → % of retrieved top-K documents that are actually relevant  
- Hit@K → Did at least one relevant document appear in top-K (Yes/No)  
- Average Similarity → Cosine similarity between query and retrieved vectors  

### Key Insight
Recall = Coverage  
Precision = Accuracy of retrieval  

---

### B. Generation Evaluation

- Factual Consistency → Are answers verifiable from retrieved sources?  
- Completeness → Does the answer fully address the query?  
- Relevance → Is the response aligned with the question?  
- Grounding → Are answers supported by retrieved context (citations)?  
- Toxicity/Bias → Is the output safe and unbiased?  

### Key Insight
Good RAG = Correct + Complete + Grounded + Safe  

---

### One-line Summary
Retrieval metrics evaluate documents, Generation metrics evaluate answers

### Quantitative Evaluation Metrics

### A. Using Synthetic Q&A Pairs
- Create dataset with:
  - Questions
  - Expected answers
  - Relevant document chunks

- Evaluate:
  - Retrieval Accuracy → Was correct document retrieved?
  - Answer Correctness → Does output match expected answer?

### Key Insight
Synthetic data enables automated evaluation without human labeling

---

### B. Embedding Similarity Check
- Use cosine similarity between:
  - Query vector and top-K retrieved vectors
  - Generated response and retrieved chunks

### Key Insight
Higher similarity = Better semantic alignment

---

### C. Automated LLM Grading (LLM-as-a-Judge)
- Use an LLM (e.g., ChatDatabricks) to evaluate responses
- Score based on:
  - Factuality
  - Completeness
  - Relevance

### Key Insight
LLMs can automate large-scale evaluation

---

### D. Human-in-the-Loop Evaluation
- Required for:
  - New domains
  - Regulated use cases
  - Edge cases

### Key Insight
Human evaluation = Highest quality but slow and costly

---

### One-line Summary
Combine synthetic, similarity, LLM, and human methods for full evaluation

### Building an Evaluation Pipeline on Databricks

### Step-by-Step

- Log Responses
  - Store each RAG response in a Delta table
  - Include → query, llm_response, retrieved_chunks, timestamp, prompt_version

- Add Ground Truth
  - Create synthetic dataset
  - Store → expected_answer, expected_chunks

- Compute Metrics (PySpark)
  - Retrieval → Recall@K, Precision@K
  - Similarity → Cosine similarity
  - Generation → BLEU, ROUGE (optional)

- Visualize Metrics
  - Use Databricks SQL / dashboards
  - Track → failures, trends, drift over time

- Manual Review
  - Flag low-confidence responses
  - Send to human review system

---

### Key Insight
Evaluation Pipeline = Logging + Metrics + Monitoring + Human Feedback

---

### One-line Summary
Track, measure, visualize, and review RAG performance continuously

### Select an LLM based on Quantitative Evaluation Metrics

### Key Metrics
- Context Precision → % of retrieved context that is relevant
- Context Recall → % of relevant context retrieved
- Faithfulness → Is answer grounded in retrieved data?
- Answer Relevance → Does answer match the question?
- Answer Correctness → Is the answer factually correct?

### Model Size vs Architecture
- Smaller models → Faster, cheaper, lower accuracy
- Larger models → Better accuracy, higher cost

### Performance vs Cost
- Consider compute cost (CPU/GPU)
- Real-time apps → Need low latency
- Batch apps → Can tolerate higher latency

### LLM-as-a-Judge (Best Practices)
- Use small rubric scales (e.g., 1–3 or 1–5)
- Provide clear examples for scoring
- Use additive scoring (e.g., 0/1/2 per metric)

### Tokens Impact
- More tokens → Better context → Better responses (but higher cost)

### Offline vs Online Evaluation
- Offline → Benchmark datasets before deployment
- Online → Real user feedback, A/B testing, drift detection

---

### Key Insight
LLM Selection = Accuracy vs Cost vs Latency trade-off

### Control LLM Costs in RAG Applications

### Cost Control Strategies
- Model Size → Use smaller models when possible to reduce cost
- Retrieval Optimization → Efficient vector search (FAISS vs Vector DB)
- Batch Processing → Use batch inference for non-real-time workloads

### Resource Tracking
- Monitor usage of CPU, GPU, and memory
- Track cost per request and overall usage

### Resource Management
- Auto-scaling → Scale resources up/down based on demand
- Resource Tagging → Track and allocate costs across teams/projects

### Key Insight
Cost Optimization = Model choice + Efficient retrieval + Smart resource usage

---

### One-line Summary
Reduce cost by optimizing model size, retrieval, and infrastructure usage
