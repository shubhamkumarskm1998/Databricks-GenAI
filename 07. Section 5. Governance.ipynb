### Section 5. Goverance

### AI System Components and Associated Risks

### 1. Data Layer (DataOps)
- Raw Data → Source systems (files, DBs, APIs)
- Data Prep → Cleaning, EDA, feature engineering
- Datasets → Train, validation, test splits

### Risks
- Poor data quality → garbage in, garbage out
- Bias in datasets → unfair model outcomes
- Data leakage → incorrect evaluation

---

### 2. Catalog (Governance Layer)
- Stores features, models, indexes
- Enables discovery, lineage, and access control

### Risks
- Lack of governance → data misuse
- Missing lineage → hard to audit/debug

---

### 3. Model Development (ModelOps)
- Algorithms → Train models (custom or foundation)
- Evaluation → Metrics, validation, testing
- Fine-tuning → Adapt pre-trained models

### Risks
- Overfitting / underfitting
- Wrong evaluation metrics
- Hallucinations in LLMs

---

### 4. Model Management
- Versioning models
- Tracking experiments
- Managing model lifecycle

### Risks
- Version mismatch
- No reproducibility
- Deployment of wrong model

---

### 5. Serving Infrastructure (DevSecOps)
- Model Serving → APIs for inference
- AI Gateway → Manage requests, routing

### Risks
- Latency and scaling issues
- Security vulnerabilities (API misuse)
- Cost overruns

---

### 6. Monitoring
- Logs → Track predictions and system behavior
- Feedback → RLHF, user signals

### Risks
- Model drift (data changes over time)
- Performance degradation
- Undetected failures

---

### 7. Cross-cutting Layers
- Governance → Policies, access control, compliance
- DataOps → Data pipelines and quality
- ModelOps → Model lifecycle management
- DevSecOps → Deployment, security, reliability

---

### Key Insight
End-to-end AI system = Data + Model + Serving + Monitoring + Governance

---

### One-line Summary
AI Pipeline Risk Areas = Data → Model → Serving → Monitoring → Governance

### Use Masking Techniques as Guardrails

### Masking Techniques
- Tokenization → Replace sensitive data with tokens
- Anonymization → Remove personally identifiable information (PII)
- Pseudonymization → Replace data with reversible identifiers

### Implementation
- Input Processing → Mask data before sending to model
- Data Storage → Store masked or encrypted data
- Inference Stage → Prevent sensitive data leakage in outputs

### Balance (Very Important)
- Ensure compliance (privacy, regulations)
- Maintain model performance (avoid over-masking)

### Key Insight
Masking = Privacy Protection without breaking model usefulness

### One-line Summary
Guardrails = Mask sensitive data across input, storage, and output

### Guardrails for Malicious User Inputs

### Guardrail Techniques
- Input Validation → Check format, length, allowed values
- Context-aware Filtering → Detect harmful or unsafe intent
- Prompt Sanitization → Remove injections or unsafe instructions

### Protection Goals
- Prevent prompt injection attacks
- Avoid harmful or inappropriate outputs

### Tools
- Llama Guard LLM → Safety model available in Databricks Marketplace

### Best Practice
- Continuously test and update guardrails for new attack patterns

### Key Insight
Guardrails = Secure input + Safe prompt + Controlled output

### Mitigation of Problematic Text in RAG Data

### Problem Areas
- Bias
- Misinformation
- Offensive / inappropriate content

### Solutions

#### Data Cleansing
- Text normalization (lowercase, cleanup)
- Remove offensive terms
- Use domain-specific stopwords

#### External Data Sources
- Use trusted and curated datasets
- Avoid noisy or unverified sources

### Key Insight
Better data = Better retrieval = Better LLM responses

---

### One-line Summary
Guardrails protect inputs, Data cleansing improves outputs

### Legal and Licensing Requirements for Data Sources

### Legal Compliance
- Understand and follow licensing terms of datasets
- Ensure data usage aligns with legal agreements
- Prevent legal risks by adhering to policies

### Dataset Licensing
- Check license type (open, restricted, commercial)
- Review datasets from sources like Databricks Marketplace
- Consult legal/compliance teams before usage

### Regular Audits
- Track data sources and their licenses
- Maintain documentation of usage rights
- Perform periodic audits to ensure compliance

### Key Insight
Compliance = Know your data + Follow license terms

### One-line Summary
Use only licensed data and audit regularly to avoid legal risks
