#### Databricks AI Platform
Databricks provides a data-centric AI platform where everything revolves around data → models → applications, all governed centrally.
**Mosaic AI** is Databricks’ GenAI platform that helps you build, customize, and deploy AI/LLM applications directly on your data.

### 1. Datasets (Data Collection & Preparation)
👉 What happens:
- Ingest, clean, transform data  
👉 Tools:
- Auto Loader → ingest streaming/batch data  
- Delta Lake → storage layer  
- Apache Spark → transformations  
- Delta Live Tables (DLT) → pipeline automation  
- Databricks Workflows → orchestration  
👉 Output:
- Clean, structured data  

### 2. Models (Use / Build Models)
👉 What happens:
- Train, fine-tune, experiment  
👉 Tools:
- MLflow → experiment tracking + model registry  
- Databricks Runtime for ML → optimized environment  
- Mosaic AI → GenAI + LLM capabilities  
- Hugging Face integration → pretrained models  
- Feature Store → feature management  
👉 Output:
- Trained / fine-tuned model  

### 3. Applications (Serving & Monitoring)
👉 What happens:
- Deploy model to production  
- Monitor performance  
👉 Tools:
- Model Serving (Databricks Serving Endpoints) → deploy APIs  
- Lakehouse Monitoring → data + model monitoring  
- Databricks SQL → dashboards & analytics  
- Vector Search → RAG-based apps  
👉 Output:
- Live AI application  

### 4. Unity Catalog (🔥 Governance Layer)
👉 What happens:
- Central control of everything  
👉 Tools:
- Unity Catalog:
  - Data governance  
  - Access control (RBAC)  
  - Lineage tracking  
👉 Covers:
- Tables  
- Models  
- Files  

### 5. Data Platform (Foundation)
👉 What happens:
- Infrastructure layer  
👉 Tools:
- Delta Lake → storage  
- Apache Spark → compute  
- Photon Engine → fast execution  
- Databricks Lakehouse Platform → unified system  

### End-to-End Flow (with tools)

- Ingest data → Auto Loader + Delta Lake  
- Transform → Apache Spark + DLT  
- Train model → MLflow + Mosaic AI  
- Deploy → Model Serving Endpoint  
- Govern → Unity Catalog

##Example 
Used Model Serving to use this end point using Open AI library

```python
from openai import OpenAI
import os

# How to get your Databricks token: https://docs.databricks.com/en/dev-tools/auth/pat.html
DATABRICKS_TOKEN = os.environ.get('DATABRICKS_TOKEN')
# Alternatively in a Databricks notebook you can use this:
# DATABRICKS_TOKEN = dbutils.notebook.entry_point.getDbutils().notebook().getContext().apiToken().get()

client = OpenAI(
    api_key=DATABRICKS_TOKEN,
    base_url="https://adb-7405611706705219.19.azuredatabricks.net/serving-endpoints"
)

response = client.chat.completions.create(
    model="databricks-meta-llama-3-1-8b-instruct",
    messages=[
        {
            "role": "user",
            "content": "What is an LLM agent?"
        }
    ],
    max_tokens=5000
)

print(response.choices[0].message.content)
```
