###Introduction to Generative AI

####What is AI?
It is science of making machine act like humans.


🧠 AI Hierarchy (Simple)

Artificial Intelligence (AI)
👉 Machines that mimic human intelligence (thinking, decision-making)

Machine Learning (ML)
👉 Subset of AI that learns from data to make predictions

Deep Learning (DL)
👉 Subset of ML using neural networks for complex tasks

Generative AI (GenAI)
👉 Subset of DL that creates new content (text, images, code)

GenAI ⊂ Deep Learning ⊂ Machine Learning ⊂ AI

⚡ Super quick memory trick

AI → Think

ML → Learn

DL → Deep patterns

GenAI → Create

🌍 Real-World Examples (Quick)

AI
👉 General intelligent systems
Examples: Siri, Alexa, Chess bots

Machine Learning (ML)
👉 Learns patterns from data
Examples: Spam filter, Netflix recommendations, Fraud detection

Deep Learning (DL)
👉 Handles complex tasks using neural networks
Examples: Face unlock, Speech-to-text, Image recognition

Generative AI (GenAI)
👉 Creates new content
Examples: ChatGPT, GitHub Copilot, DALL·E
