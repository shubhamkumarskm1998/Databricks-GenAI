### Agentic AI

### What is Agentic AI?
Agentic AI refers to systems where an AI agent can plan, reason, and take actions autonomously using tools and data.

---

### Flow

User  
  ↓  
AI Agent  
  ↓  
LLM (Reasoning / Decision Making)  
  ↓  
Tools / Actions (APIs, functions, workflows)  
  ↓  
Databases (Structured Data)  
Vector DB (Embeddings / Context)  
  ↓  
Feedback Loop → Improves future responses  

---

### Components

- AI Agent → Orchestrates the workflow  
- LLM → Thinks, reasons, decides next step  
- Tools / Actions → Execute tasks (API calls, DB queries)  
- Database → Structured data storage  
- Vector DB → Semantic search (RAG)  
- Feedback Loop → Learning and improvement  

---

### Key Insight
Agent = LLM + Tools + Memory + Actions

---

### Why Agentic AI?

- Automates multi-step tasks  
- Enables decision-making (not just answering)  
- Integrates with real-world systems  

---

### Example

User: "Book me a flight"  
→ Agent searches flights  
→ Calls booking API  
→ Confirms booking  

---

### One-line Summary
Agentic AI = LLM that can think, act, and interact with tools autonomously

### Agent Brick – Declarative Agents

### What are Agent Bricks?

Agent Bricks are pre-built templates in Databricks to quickly create AI agents using a declarative approach (configuration instead of coding).

---

### Key Idea

Declarative = Define *what* the agent should do  
Not imperative coding = *how* to do it

---

### Available Templates

- Information Extraction  
  → Extract structured data from documents (JSON output)

- Knowledge Assistant  
  → Build RAG-based chatbot using documents

- Multi-Agent Supervisor  
  → Orchestrate multiple agents working together

- Custom LLM  
  → Create domain-specific AI applications

---

### How It Works

User selects template  
→ Configure data sources (Delta / Vector DB)  
→ Define tools & prompts  
→ Deploy agent  

---

### Benefits

- Faster development (no heavy coding)
- Standardized architecture
- Easy integration with Databricks ecosystem
- Built-in best practices (RAG, tools, evaluation)

---

### Where It Fits

Agent Bricks → Part of **Build Agents** stage in Agentic AI lifecycle

---

### Key Insight
Agent Bricks = Low-code / no-code way to build production-ready AI agents

---

### One-line Summary
Agent Bricks help you quickly build AI agents using templates instead of writing everything from scratch

### How to Address GenAI Challenges

---

### 1. Data Bias

- Assess data slices → Identify bias across groups
- Update data → Improve diversity and representation

---

### 2. Toxic / Discriminatory Output

- Post-hoc tools → Detect harmful outputs after generation
- Guardrails → Prevent unsafe or inappropriate responses

---

### 3. Misinformation

- Fine-tune model → Improve factual accuracy
- Use better datasets → High-quality, verified data
- RAG → Retrieve trusted external knowledge

---

### 4. Malicious Use

- Regulation → Policies and governance controls
- Access control → Restrict usage
- Monitoring → Detect misuse patterns
- Rate limiting → Prevent abuse

---

### Key Insight
Challenges in GenAI are solved using a combination of:
Data + Models + Guardrails + Governance

---

### One-line Summary
Handle GenAI risks using bias mitigation, guardrails, fine-tuning, and governance controls

### Mitigate Risk in GenAI Systems

---

### Key Areas

- Data Licensing  
  → Ensure legal usage of datasets

- Data Accreditation  
  → Use trusted, verified data sources

- Prompt Safety  
  → Prevent harmful or malicious prompts

- Guardrails  
  → Enforce safe and compliant outputs

- Evaluate LLM Quality  
  → Measure accuracy, relevance, and safety

- Secure the System  
  → Access control, encryption, and monitoring

- Evaluate System Quality  
  → End-to-end testing (RAG + LLM + pipelines)

---

### Key Insight
Risk mitigation requires coverage across:
Data + Model + Prompt + System

---

### One-line Summary
Mitigate GenAI risk using licensing, safety controls, evaluation, and system security
