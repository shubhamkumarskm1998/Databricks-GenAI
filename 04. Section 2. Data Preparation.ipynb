### Understanding Retrieval-Augmented Generation (RAG) in Databricks
### Why RAG is Needed
- LLMs (DBRX, GPT-4, MPT via Foundation Model APIs) are:
  - trained on public data  
  - static (no real-time updates)  
  - unaware of enterprise data in Delta tables  
  - prone to hallucinations  
- RAG uses Databricks data + retrieval to improve accuracy  
### RAG Flow (Databricks)
- Data Preparation → ingest data using Auto Loader + store in Delta Lake  
- Indexing → create embeddings using Foundation Model APIs / embedding models  
- Storage → store embeddings in Mosaic AI Vector Search index  
- Retrieval → query vector index to get relevant chunks  
- LLM Inference → use Model Serving / Foundation Model APIs to generate response  
### Key Databricks Components
- Delta Lake → store raw + processed data  
- Embedding Models → convert text to vectors  
- Mosaic AI Vector Search → retrieval engine  
- Retriever → fetch top-k relevant chunks  
- Model Serving / FM APIs → generate final response  
- Unity Catalog → governance and access control  
### Example (Databricks)
- User query → “How to track billing?”  
- Vector Search → retrieves relevant document chunks  
- LLM (served via endpoint) → generates contextual answer  
### Benefits in Databricks
- Works with enterprise data in Delta  
- Secure via Unity Catalog  
- Scalable with Spark + Lakehouse  
- Reduced hallucination with grounded responses  
### One-line Summary
**Databricks RAG = Delta Lake + Vector Search + LLM Serving → accurate, enterprise-ready GenAI**

### What is Retrieval-Augmented Generation (RAG)
- RAG combines:
  - Retrieval → fetch relevant data using vector search over embeddings  
  - Generation → LLM generates response using retrieved context  
### Formal Definition
- RAG → hybrid approach where LLM uses dynamically retrieved documents to improve accuracy, reduce hallucinations, and use private/recent data  
- Works at inference time (no retraining like fine-tuning)  
### Core Components of RAG
- User Query → input question  
- Embedding Model → converts text to vectors  
- Vector Search (Mosaic AI Vector Search) → similarity search  
- Retriever → fetch top-K relevant chunks  
- Prompt Constructor → combines query + retrieved context  
- LLM (Model Serving / FM APIs) → generates final response  
### Traditional LLM vs RAG
- Data Access:
  - Traditional → fixed (training data)  
  - RAG → dynamic (external sources)  
- Enterprise Knowledge:
  - Traditional → not accessible  
  - RAG → easily integrated  
- Updates:
  - Traditional → requires fine-tuning  
  - RAG → real-time retrieval  
- Hallucination:
  - Traditional → higher  
  - RAG → reduced  
- Explainability:
  - Traditional → difficult  
  - RAG → traceable via retrieved documents  
### One-line Summary
**RAG = Retrieval (Vector Search) + Generation (LLM) → accurate, real-time, enterprise-ready AI**

### Building a RAG System (Databricks)
### Steps
- Import required LangChain + Databricks libraries  
- Connect to Vector Search index  
- Connect to LLM (Model Serving endpoint)  
- Create RAG chain (Retriever + LLM)  
- Run query and get response  
### Components
- Vector Search → retrieves relevant document chunks  
- LLM → generates final answer  
- RetrievalQA → combines retrieval + generation  
### One-line Summary
**RAG = Vector Search + LLM → context-aware answer generation**

```python
# Imports
from langchain.vectorstores import DatabricksVectorSearch
from langchain.embeddings import DatabricksEmbeddings
from langchain.chains import RetrievalQA
from langchain.chat_models import ChatDatabricks

# Vector Search setup
vector_search = DatabricksVectorSearch(
    index_name="insurance_policies_index")

# Connect to LLM
llm = ChatDatabricks(
    endpoint="databricks-meta-llama-3-1-70b-instruct")

# Build RAG chain
rag_chain = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=vector_search.as_retriever(),
    return_source_documents=True)

# Query
query = "What is the policy coverage for maternity benefits?"
response = rag_chain.run(query)

print(response)
```

### Chunking Strategies for Semantic Understanding
### What is Chunking
- Breaking large documents into smaller semantic units for embedding and retrieval  
### A Good Chunk
- Self-contained  
- Topically consistent  
- Fits within LLM context window  
### Chunking Techniques
- Fixed-length → split every N tokens (e.g., 500)  
- Sentence-aware → split by sentence boundaries (better semantics)  
- Header-based → split using section headings (e.g., Coverage, Terms)  
- Sliding window → overlapping chunks (e.g., 500 tokens with 100 overlap)  
### Key Insight
- Avoid blind fixed chunking → prefer semantic + structured chunking  
### One-line Summary
**Good chunking = better retrieval = better RAG performance**

### Retrieval Evaluation Metrics

- Context Precision → % of retrieved chunks that are relevant  
- Context Recall → % of relevant chunks successfully retrieved  
- Faithfulness → whether answer is grounded in retrieved context  
- Answer Relevance → how well answer addresses the query  
- Answer Correctness → factual and semantic accuracy  

---

### Evaluation Tools and Methods
- MLflow → evaluate retrievers + LLMs, batch experiments, scalable tracking  
- LLM-as-a-Judge → LLM evaluates another LLM using scoring criteria  
- Task-specific Metrics → BLEU, ROUGE  

---

### Key Metrics (with Definitions)
- BLEU (Bilingual Evaluation Understudy) → measures overlap between generated text and reference text (used in translation)  
- ROUGE (Recall-Oriented Understudy for Gisting Evaluation) → measures overlap of n-grams, recall-focused (used in summarization)  

- Precision@K → % of top-K retrieved results that are relevant  
- Recall@K → % of all relevant results found in top-K  
- MRR (Mean Reciprocal Rank) → inverse rank of first relevant result (higher = better ranking)  
- nDCG (Normalized Discounted Cumulative Gain) → ranking quality based on position and relevance  
- Hit Rate → whether at least one relevant result appears in top-K  

- Latency → time taken to generate response  
- Token Usage → number of tokens used (cost metric)  

---

### Offline vs Online Evaluation
- Offline Evaluation → before deployment  
  - Metrics → BLEU, ROUGE, Precision@K, Recall@K, nDCG  

- Online Evaluation → after deployment  
  - Metrics → CTR (Click-through rate), user feedback, A/B testing, latency, cost

### max_tokens and temperature (ChatDatabricks)

### temperature → controls randomness
- Range → 0 to 1 (sometimes >1)
- Low (0–0.3) → deterministic, factual, consistent  
- Medium (0.4–0.7) → balanced  
- High (0.8–1) → creative, diverse, less predictable  

👉 Example:
- temperature = 0.1 → best for RAG, Q&A, factual tasks  

---

### max_tokens → controls output length
- Maximum number of tokens generated in response  
- 1 token ≈ 0.75 words (approx)  

👉 Example:
- max_tokens = 250 → limits response size  

---

### In your code
- temperature = 0.1 → accurate, low randomness  
- max_tokens = 250 → short, controlled response  

---

### Best Practice (Exam)
- RAG / QA → low temperature (0–0.3)  
- Creative writing → high temperature (0.7–1)  
- Control cost → reduce max_tokens  

---

### One-line Summary
**temperature = randomness | max_tokens = response length**
